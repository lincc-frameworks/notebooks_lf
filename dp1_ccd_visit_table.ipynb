This notebook queries butler to get the CcdVisit table for DP1. 
Schema can be found here https://sdm-schemas.lsst.io/dp1.html#CcdVisit

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
import pandas as pd
from tqdm import tqdm
from upath import UPath

from lsst.daf.butler import Butler
```

```python
#From tutorial notebook 104_2

butler = Butler("dp1", collections="LSSTComCam/DP1")
```

```python
visit_refs = butler.query_datasets("visit_detector_table")

print(len(visit_refs))
```

```python
ref = visit_refs[0]
uri = butler.getURI(ref)
path = UPath(uri.geturl())
path
```

```python
data = path.read_bytes()
with UPath("ccd_visits.parquet").open("wb") as local:
    local.write(data)
```

```python
pd.read_parquet('ccd_visits.parquet', dtype_backend='pyarrow')
```
