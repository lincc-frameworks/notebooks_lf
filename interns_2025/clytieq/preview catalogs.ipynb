```python
def print_crossmatched_cols(path, xmatch, ret=False):
    xmatched = lsdb.open_catalog(path)
    cols = xmatched.columns
    cats = {'xmatch': [], 'catalog': []}
    suf1, suf2 = xmatch
    for col in cols:
        if xmatch is not None:
            if col[0] == "_":
                cats['xmatch'].append(col)
            else:
                # if suf1 in col:
                # colsplit = col.split("_")
                # suffix = colsplit[-1]
                # col = "_".join(colsplit[:-1])
                if suffix in cats:
                    cats[suffix].append(col)
                else:
                    cats[suffix] = [col]
        else:
            cats['catalog'].append(col)
    for cat in cats:
        if (len(cats[cat]) > 0):
            print(f"{cat}:", ", ".join(cats[cat]))

    if ret:
        return xmatched
    else:
        return xmatched.head()
```

```python
from astropy.coordinates import SkyCoord
import astropy.units as u


# min_ra, max_ra = 251.45, 251.5
# min_dec, max_dec = 52.05, 52.1

# Example RA and Dec (in degrees)
# min
# ra = 251.45  
# dec = 52.05  

# center
ra = 251.475
dec = 52.075

# max
# ra = 251.5
# dec = 52.1

# Create SkyCoord object
sky_coord = SkyCoord(ra=ra*u.degree, dec=dec*u.degree, frame='icrs')

# Convert to Galactic coordinates
galactic = sky_coord.galactic

# Access l and b
l = galactic.l.deg
b = galactic.b.deg

print(f"Galactic Longitude (l): {l} degrees")
print(f"Galactic Latitude (b): {b} degrees")
print(f"{ra}, {dec}")
print(f"{l}, {b}")
```
