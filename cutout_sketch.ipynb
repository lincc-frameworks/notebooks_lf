```python
from nested_pandas.datasets import generate_data
import pandas as pd
import numpy as np
```

```python
nf = generate_data(5,1000)
nf

nf["cutout_x"] = 150
nf["cutout_y"] = 150

# generate cutout arrays

#cutouts = np.random.rand(5,150,150)
for band in ["u","g","r","i","z","y"]:
    
    cutout_data = np.random.rand(5, 22500)
    cutout_var = np.random.rand(5, 22500)
    cutout_mask = np.random.rand(5, 22500)
    cdata_series = pd.Series(list(cutout_data))
    cvar_series = pd.Series(list(cutout_var))
    cmask_series = pd.Series(list(cutout_mask))
    nf[f"{band}_data"] = cdata_series
    nf[f"{band}_variance"] = cvar_series
    nf[f"{band}_mask"] = cmask_series
nf
#nf.nest_lists(columns=["data", "variance", "mask"], name="cutouts")
```

## Approach 1: "Ugly"

```python
nf_ugly = nf
for col in nf.columns[5:]:
    nf_ugly = nf_ugly.nest_lists(columns=[col], name=col)
nf_ugly
```

## Approach 2: Band nested cols

```python

nf_band = nf

for band in ["u","g","r","i","z","y"]:
    band_cols = [col for col in nf_band.columns if f"{band}_" in col]
    print(band_cols)
    nf_band = nf_band.nest_lists(columns=band_cols, name=band)
nf_band
```

## Approach 3: By image layer

```python
nf_layer = nf

for layer in ["data","variance","mask"]:
    layer_cols = [col for col in nf_layer.columns if f"_{layer}" in col]
    print(layer_cols)
    nf_layer = nf_layer.nest_lists(columns=layer_cols, name=layer)
nf_layer
```
