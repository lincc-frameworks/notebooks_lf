## Catalog partition slicer

We can now use the square brackets accessor to obtain a new catalog with a subset of partitions. It was implemented to function in the same way as it does for Python lists. We can use a single integer to access a single partition, but we can also use slices with the `[start:stop:step]` notation.

```python
import lsdb
```

```python
gaia = lsdb.read_hipscat("/data3/epyc/data3/hipscat/catalogs/gaia_dr3/gaia")
```

```python
gaia.partitions
```

```python
gaia.partitions[2] # Partition 2
```

```python
gaia.partitions[2:6] # Partitions 2, 3, 4 and 5
```

```python
gaia.partitions[2:10:3] # Partitions 2, 5 and 8
```
