```python
import lsdb
```

```python
ztf = lsdb.read_hipscat('https://data.lsdb.io/unstable/ztf/ztf_dr14/')
ztf
```

```python
ztf.columns
```

```python
ztf = ztf[["ra", "dec", "ps1_objid", "mean_mag_r"]]
```

```python
ps1 = lsdb.read_hipscat('s3://stpubdata/panstarrs/ps1/public/hipscat/otmo', storage_options={'anon': True})
```

```python
ps1
```

```python
for c in ps1.columns:
    print(c)
```

```python
ps1 = ps1[["objID", "raMean", "decMean", "rMeanApMag"]]
```

```python
ztf_cone = ztf.cone_search(10, 10, 100)
ztf_cone
```

```python
ztf_source = lsdb.read_hipscat('https://data.lsdb.io/unstable/ztf/ztf_zource/')
```

```python
ztf_source.columns
```

```python
ztf_source = ztf_source[["ps1_objid", "band", "mag",  "mjd"]]
```

```python
ps1_source = lsdb.read_hipscat('s3://stpubdata/panstarrs/ps1/public/hipscat/detection', storage_options={'anon': True},
                               columns=["objID","ra","dec","zp","psfFlux","psfFluxErr","filterID","obsTime"]
                               )
```

```python
ztf_x_ps = ztf_cone.crossmatch(ps1)
ztf_x_ps
```

```python
ztf_x_ps_with_lcs = ztf_x_ps.join_nested(
        ztf_source , left_on = "ps1_objid_ztf_dr14", right_on="ps1_objid",nested_column_name="ztf_lc"
    ).join_nested(
        ps1_source , left_on = "objID_otmo", right_on="objID", nested_column_name="ps_lc"
    )
```

```python
computed_table = ztf_x_ps_with_lcs.compute()
computed_table
```

```python
computed_table.iloc[0]["ztf_lc"]
```

```python
import matplotlib.pyplot as plt
import numpy as np

for ii in range(5):


    ## Get light curves:
    ps_lightcurve = computed_table.ps_lc.values[ii]
    ztf_lightcurve = computed_table.ztf_lc.values[ii]


    ## PANSTARRS == 
    ps_time = ps_lightcurve["obsTime"]
    ps_flux = ps_lightcurve["psfFlux"]

    # PSFflux to AB mag
    ps_mag = -2.5*np.log10(ps_flux) + 8.9

    plt.plot(ps_time , ps_mag , "s", markersize=3 , label="Pan-STARRS")


    ## ZTF == 
    ztf_time = ztf_lightcurve["mjd"]
    ztf_mag = ztf_lightcurve["mag"]
    plt.plot(ztf_time , ztf_mag , "d", markersize=3 , label="ZTF")

    plt.xlabel("MJD")
    #plt.ylabel("Flux")
    #plt.ylabel("Mag")
    plt.ylabel("mJy")
    plt.legend(bbox_to_anchor=(1,1))
    plt.gca().invert_yaxis()

    plt.show()
```

```python
ztf_x_ps_with_lcs_ztf_r = ztf_x_ps_with_lcs.query("ztf_lc.band == 'r'")
```

```python
computed_filtered_table = ztf_x_ps_with_lcs_ztf_r.compute()
```

```python
computed_filtered_table
```

```python
for ii in range(5):


    ## Get light curves:
    ps_lightcurve = computed_filtered_table.ps_lc.values[ii]
    ztf_lightcurve = computed_filtered_table.ztf_lc.values[ii]
    
    if ztf_lightcurve is None:
        continue


    ## PANSTARRS == 
    ps_time = ps_lightcurve["obsTime"]
    ps_flux = ps_lightcurve["psfFlux"]

    # PSFflux to AB mag
    ps_mag = -2.5*np.log10(ps_flux) + 8.9

    plt.plot(ps_time , ps_mag , "s", markersize=3 , label="Pan-STARRS")


    ## ZTF == 
    ztf_time = ztf_lightcurve["mjd"]
    ztf_mag = ztf_lightcurve["mag"]
    plt.plot(ztf_time , ztf_mag , "d", markersize=3 , label="ZTF")

    plt.xlabel("MJD")
    #plt.ylabel("Flux")
    #plt.ylabel("Mag")
    plt.ylabel("mJy")
    plt.legend(bbox_to_anchor=(1,1))
    plt.gca().invert_yaxis()

    plt.show()
```
