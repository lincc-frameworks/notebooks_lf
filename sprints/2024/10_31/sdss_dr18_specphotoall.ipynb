# SDSS DR18 SpecPhotoAll

This notebook demonstrates the importing process for SDSS DR18 SpecPhotoAll.

```python
import lsdb
import pandas as pd
from dask.distributed import Client
from hats_import.pipeline import pipeline_with_client
from hats_import.catalog.arguments import ImportArguments
from hats_import.catalog.file_readers import CsvReader
```

```python
file_path = "/epyc/data3/hats/tmp/sdss_dr18_specphotoall/SpecPhotoAll.csv"
```

Load the column names and types from a side file:

```python
type_frame = pd.read_csv("/epyc/data3/hats/tmp/sdss_dr18_specphotoall/schema.csv")
type_map = dict(zip(type_frame["name"], type_frame["type"]))
type_map
```

```python
args = ImportArguments(
    output_artifact_name="sdss_dr18_specphotoall",
    input_file_list=[file_path],
    file_reader=CsvReader(
        header=None,
        skiprows=[0],
        column_names=type_frame["name"].values.tolist(),
        type_map=type_map,
        chunksize=250_000,
    ),
    ra_column="ra",
    dec_column="dec",
    output_path="/epyc/data3/hats/catalogs",
)
```

```python
with Client(n_workers=64) as client:
    pipeline_with_client(args, client)
```

Gaia does not provide us with (accurate) redshifts for its objects so we're interested in crossmatching the two catalogs.

```python
sdss_dr18_specphotoall = lsdb.read_hats("/epyc/data3/hats/catalogs/sdss_dr18_specphotoall")
sdss_dr18_specphotoall
```

```python
sdss_dr18_specphotoall.head(10)
```

```python
gaia = lsdb.read_hats(
    'https://data.lsdb.io/hats/gaia_dr3/gaia', 
    margin_cache='https://data.lsdb.io/hats/gaia_dr3/gaia_10arcs'
)
```

```python
sdss_x_gaia = sdss_dr18_specphotoall.crossmatch(gaia)
sdss_x_gaia
```

```python
with Client(n_workers=32):
    result = sdss_x_gaia.head(10)
```
