```python
import hats
from astropy.coordinates import Angle, SkyCoord
```

```python
catalog = hats.read_hats('https://data.lsdb.io/hats/wise/allwise/')
```

```python
from astropy.visualization.wcsaxes.frame import EllipticalFrame
import matplotlib.pyplot as plt
from matplotlib.colors import LogNorm
from mocpy import WCS

wcs = WCS(
    fig=plt.gcf(),
    fov=(Angle(320, 'deg'), Angle(160, 'deg')),
    coordsys='galactic',
    center=SkyCoord(l=0, b=0, unit='deg', frame='galactic'),
)

fig, wcs = hats.inspection.plot_points(
    catalog,
    norm=LogNorm(),
    wcs=wcs.w,
    frame_class=EllipticalFrame
)

wcs.set_xlabel('l')
wcs.set_ylabel('b')
```

```python
import numpy as np
from matplotlib.colors import LogNorm

fig, wcs = hats.inspection.plot_points(
    catalog,
    norm=LogNorm(),
)
wcs.scatter(np.r_[0:360:100j], 30 + 5 * np.sin(np.r_[0:360:100j]), transform=wcs.get_transform('galactic'), s=300,
            edgecolor='orange', facecolor='none')
wcs.scatter(np.r_[0:360:100j], -30 + 5 * np.sin(np.r_[0:360:100j]), transform=wcs.get_transform('galactic'), s=300,
            edgecolor='purple', facecolor='none', marker='^')

wcs.set_xlabel('l')
wcs.set_ylabel('b')
```
