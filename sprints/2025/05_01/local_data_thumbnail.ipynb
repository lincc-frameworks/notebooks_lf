# Data Thumbnail Creation

This notebook shows a method of creating the data thumbnail for an existing catalog.

```python
import lsdb
```

Create a small version of Gaia DR3 for testing this approach, before trying it on a full catalog.

```python
gaia3_sm = lsdb.read_hats(
    '/data3/epyc/data3/hats/catalogs/gaia_dr3/gaia',
    search_filter=lsdb.ConeSearch(ra=280, dec=-20, radius_arcsec=2*3600)
)
```

Check how many partitions exist in this smaller catalog.
The data thumbnail should have one row per partition.

```python
gaia3_sm.npartitions
```

Use map_partitions to take the top row from every partition.

```python
top_rows = gaia3_sm.map_partitions(lambda df: df.head(1))
top_rows
```

```python
%%time
top_rows.compute()
```

Now for the whole catalog!

```python
gaia3 = lsdb.read_hats(
    '/data3/epyc/data3/hats/catalogs/gaia_dr3/gaia',
)
```

```python
top_rows_all = gaia3.map_partitions(lambda df: df.head(1))
top_rows_all
```

```python
import pyarrow.parquet as pq
```

```python
top_rows
```

Write this thumbnail to a *single* Parquet file.

```python
type(top_rows)
```

```python
top_rows_df = top_rows.compute()
```

```python
type(top_rows_df)
```

```python
top_rows_df.to_parquet("data_thumbnail.parquet")
```

```python
!ls -lh *.parquet
```

Now for the whole catalog.

```python
%%time
top_rows_all_df = top_rows_all.compute()
top_rows_all_df
```

```python
top_rows_all_df.to_parquet("data_thumbnail.parquet")
```
