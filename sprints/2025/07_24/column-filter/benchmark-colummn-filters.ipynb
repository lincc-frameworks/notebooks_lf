```python
import lsdb
from upath import UPath

columns_ = [
    ['_healpix_29',],
    ['_healpix_29', 'ra', 'dec',],
    ['_healpix_9', '_healpix_19', '_healpix_29', 'ra', 'dec', 'object_id', 'class_phz_classification'],
    None,
    'all',
]
region = lsdb.PixelSearch([(6, 16045)])

block_sizes = [16 << 10, 256 << 10, 1 << 20, 5 << 20]

for columns in columns_:
    for block_size in block_sizes:
        print(f'Columns: {columns}, Block size: {block_size}')
        catalog = lsdb.open_catalog(
            UPath('s3://nasa-irsa-euclid-q1/contributed/q1/merged_objects/hats', anon=True, default_block_size=block_size),
            columns=columns,
            search_filter=region,
        )
        %time catalog.compute()
        print()
```
