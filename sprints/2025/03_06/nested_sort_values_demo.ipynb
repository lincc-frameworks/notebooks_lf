# Demo: Nesting Sorting

## Nested-Pandas (Added in v0.3.7)

```python
# Generate some toy data
ndf = npd.datasets.generate_data(5,30)
ndf
```

```python
ndf.sort_values?
```

```python
# Normal sort_values behavior
ndf.sort_values(by="a", ascending=False)
```

```python
# Applying to Nested Layers works as you'd expect
ndf_band_sorted = ndf.sort_values(by=["nested.band"])
ndf_band_sorted
```

```python
ndf_band_sorted.iloc[0]["nested"]
```

```python
# Some cool things like multi-tiered sorting

ndf_band_msorted = ndf.sort_values(by=["nested.band", "nested.t"])
ndf_band_msorted
```

```python
ndf_band_msorted.iloc[0]["nested"]
```

## Nested-Dask (On Main, but not part of a release yet)

```python
ndd = nd.datasets.generate_data(100,50, npartitions=5)
ndd
```

```python
ndd.sort_values?
```

```python
# Normal Behavior
ndd.sort_values(by="a") # sorts are a bigger deal in dask, we lose divisions obviously!
```

```python
# Nested-Dask
sorted_ndd = ndd.sort_values(by=["nested.band"])
sorted_ndd # nested sorting is a contained problem, so divisions are preserved
```

```python
sorted_ndd.head(5)
```

```python
sorted_ndd.head(5).iloc[0]["nested"]
```
