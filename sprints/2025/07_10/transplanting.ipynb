# Possible AGN problem 3 - run a function across AGN lightcurves

## Computing `drw_fit` in memory

At this point we have `m_x_dp1_obj` in memory, and it's easy to examine and process it.

```python
from nested_pandas.utils import count_nested

m_x_dp1_obj_ct = count_nested(m_x_dp1_obj_c, "objectForcedSource_object_lc")
long_light = m_x_dp1_obj_ct.query("n_objectForcedSource_object_lc > 200")
long_light
```

```python
from eztao.ts import drw_fit
```

```python
%%time
def do_fit(lc_nest_in):
    lc_nest = lc_nest_in.dropna()
    coeffs = drw_fit(
        lc_nest["midpointMjdTai"].to_numpy(),
        lc_nest["psfMag"].to_numpy(),
        lc_nest["psfMagErr"].to_numpy())
    drw_0, drw_1 = coeffs
    return { "drw_0": drw_0, "drw_1": drw_1 }

fits = long_light.reduce(do_fit, "objectForcedSource_object_lc")
fits
```

## Computing `drw_fit` on the catalog

So far so good!  But suppose that we had a lot more than 478 rows and we wanted to leverage the cluster.

Now we want to transplant our in-memory computations confidently to the catalog.  The big problem here is "confidently".

```python
# Step 1: count the number of elements in each nest and only keep those with > 200 points.

from nested_pandas.utils import count_nested

def count_points(pts):
    # Asked to count `objectForcedSource_object_lc`, this will add a column called `n_objectForcedSource_object_lc`
    return count_nested(pts, "objectForcedSource_object_lc")

counted_cat = m_x_dp1_obj.map_partitions(count_points)
long_light_cat = counted_cat.query("n_objectForcedSource_object_lc > 200")
```

### Transplant direct to `lsdb.catalog.Catalog.reduce`?

```python
# Use the previous function directly.
# Hideous `meta=` required.
fit_cat = long_light_cat.reduce(do_fit, "objectForcedSource_object_lc", meta={"drw_0": float, "drw_1":float})
fit_cat
```

#### This actually worked!

```python
fit_cat.compute()
```
