## Progress bar in Dask

Small investigation on how to provide an optional progress bar to users. Useful during potentially-long compute requests ([#55](https://github.com/astronomy-commons/lsdb/issues/55)).

```python
import lsdb
import dask.diagnostics
import dask.distributed
```

```python
cone_search = lsdb.ConeSearch(ra=0, dec=0, radius_arcsec=2*3600)
twomass = lsdb.open_catalog('https://data.lsdb.io/hats/two_mass', search_filter=cone_search)
delve = lsdb.open_catalog('https://data.lsdb.io/hats/delve_dr2', search_filter=cone_search)
```

```python
xmatch = twomass.crossmatch(delve)
xmatch
```

```python
def check_client():
    try:
        client = dask.distributed.get_client()
        print("Using distributed scheduler")
    except ValueError:
        print("Not using distributed scheduler")
```

### Default (local) scheduler

```python
check_client()
```

```python
with dask.diagnostics.ProgressBar():
    result = xmatch.compute()
result
```

### Distributed scheduler

```python
client = dask.distributed.Client(n_workers=2, threads_per_worker=1)
client
```

```python
check_client()
```

This works and shows all task names (similar to what happens in the dashboard) because it detects we're running on a Jupyter notebook:

```python
# start computation in the background
result = xmatch._ddf.persist()
# watch progress (non-blocking, grouped by task keys)
dask.distributed.progress(result)
```

```python
# release previously persisted result
client.cancel(result)
```

The task names are not very helpful, we can also just get a single progress bar:

```python
# release persisted result
result = xmatch._ddf.persist()
# watch progress (blocking, single bar)
dask.distributed.progress(result, notebook=False)
```

In every case we need to call persist() first which has some pros/cons (see [#613](https://github.com/astronomy-commons/lsdb/issues/613)).

```python
client.close()
```
