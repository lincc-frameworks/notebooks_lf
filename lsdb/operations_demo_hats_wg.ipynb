## Implementation Complexity & Current LSDB Challenges

We have done a lot of custom work within the arrow/pandas/dask stack, and a significant portion of our development work has been about wiring these things up correctly to use Dask within the more mainline path of the DataFrames API.

This has led us to a lot of performance related issues, where we weren't sure why something was using so much memory, or why a graph size for a given workflow was so large. It's not to say that Dask inherently was responsible for these issues, and likely it's more about a tension between the relatively focused scope of what LSDB intends to provide, and the work we've needed to do to make a generalist package like Dask work for us.

## Exploring Custom LSDB Graphs - "Operations"

Dask supports custom graphs, and we recently started doing some exploratory work on what that might look like for LSDB. We've had some really promising early results/progress, which has spun up into a more focused effort to fully implement them as the new graph construction backend.

The `Operation` is one of two new foundational classes introduced in the LSDB Custom Graphs work. It is responsible for tracking the construction of a task graph, as well as tracking the lazy meta of a given workflow.

## Scope of Operations

Generalized fully, Operations would likely retread much of the same development that Dask itself has made with expressions and more generally graph construction. We are leveraging the fact that LSDB has a much smaller scope of supported functionality to keep Operations local to the things we care about:

### Crossmatching

By matching pixels through spatial information, we know the full optimal task list a priori.

```python
gaia = lsdb.open_catalog('https://data.lsdb.io/hats/gaia_dr3')
tess = lsdb.open_catalog('https://data.lsdb.io/hats/tess/tic')
gaia.crossmatch(tess, suffix_method="overlapping_columns")
```

### Map-Partitions

Essentially the rest of the API is map-partitions operations over pandas-level operations. Even things like joins, which may involve full searches between all partitions are restricted to spatial joins.

## Early Findings

```python
import lsdb
import sys

# Load Catalogs
gaia = lsdb.open_catalog('https://data.lsdb.io/hats/gaia_dr3')
tess = lsdb.open_catalog('https://data.lsdb.io/hats/tess/tic')
```

### Custom Graphs

```python
%%time

# Graph Building Time -- Custom Graphs
xmatch = gaia.crossmatch(tess, suffix_method='overlapping_columns')
print(f"Number of Graph Tasks: {len(xmatch._operation.build().graph)}")
```

```python
%%time
# Graph Memory Size -- Custom Graphs
graph=xmatch._operation.build().graph
graph_size = 0
for key in graph.keys():
    graph_size+=sys.getsizeof(graph[key])
print(f"Graph Size (MB): {graph_size/10**6}")
```

### Current LSDB

```python
%%time

# Graph Building Time -- Current LSDB
xmatch = gaia.crossmatch(tess, suffix_method='overlapping_columns')
print(f"Number of Graph Tasks: {len(xmatch._ddf.dask)}")
```

```python
%%time
# Graph Memory Size -- Current LSDB
graph=xmatch._ddf.dask
graph_size = 0
for key in graph.keys():
    graph_size+=sys.getsizeof(graph[key])
print(f"Graph Size (MB): {graph_size/10**6}")
```

## Summary

Custom graph creation looks to be a very promising direction for LSDB, allowing full control of our input graphs while still using Dask for execution. We're able to reduce our graph sizes by a factor of a few, which helps for our scalability objectives where large graphs can be a real failure mode.

Our approach feels very similar to what Dask Expressions aims to do, as we even aim to do things like query optimization down the road. It's very possible that we could approach this as something where we create custom expressions for our use cases, but at the same time, creating custom graphs for Dask is pretty intuitive and offers a really strong sense of control.
